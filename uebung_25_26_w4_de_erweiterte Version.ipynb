- 90‚Äì100 ‚Üí ‚ÄûAusgezeichnet‚Äú
- 75‚Äì89 ‚Üí ‚ÄûGut‚Äú
- 60‚Äì74 ‚Üí ‚ÄûBefriedigend‚Äú

```python
# Bsp. einer FUNKTION:

def add_numbers(a,b): # Deklaration, Funktionsname, Parameter
    result = a + b    # Funktionsk√∂rper
    return result     # R√ºckgabewert 
sum = add_numbers(5,3) # Aufruf der Funktion
print(sum)             # Ausgabe des Aufrufs


# ohne "sum" (k√ºrzer):
def add_numbers(a,b):
    result = a + b
    return result

print(add_numbers(5,3))
```

```python
# L√ñSUNG 5.1: FUNKTION mit Loop:

def bewerte(punkte):
    if 90 <= punkte <= 100:
        return "Ausgezeichnet"
    elif 75 <= punkte <= 89:
        return "Gut"
    elif 60 <= punkte <= 74:
        return "Befriedigend"
    elif 40 <= punkte <= 59:
        return "Ausreichend"
    elif 0 <= punkte <= 39:
        return "Verbesserungsw√ºrdig"
    else:
        return "Ung√ºltige Punktzahl"

# Tests
for p in [95, 80, 65, 50, 12, -1, 105]:
    print(p, "‚Üí", bewerte(p))
```

<b>Aufgabe 5.2 (mit KI):</b> Lassen Sie sich von der KI eine kurze Funktionssammlung (3 kleine, kommentierte Funktionen)
f√ºr typische Mini-Aufgaben erstellen (z.&nbsp;B. <i>ist_gerade</i>, <i>max2</i>, <i>ist_vokal</i>). F√ºgen Sie Code und 2 eigene Tests hier ein.

```python
# Aufgabe 5.2 ‚Äì weiterentwickelte L√∂sung mit input-Fenster, f√ºr integer values:

def ist_gerade(n):
    """
    Gibt True zur√ºck, wenn n durch 2 teilbar ist.
    Erwartet eine ganze Zahl (int).
    """
    return n % 2 == 0


def max2(a, b):
    """
    Gibt die gr√∂√üere von zwei ganzen Zahlen zur√ºck.
    """
    return a if a >= b else b


def ist_vokal(c):
    """
    Pr√ºft, ob ein Zeichen ein deutscher Vokal ist.
    """
    return c.lower() in ["a", "e", "i", "o", "u", "√§", "√∂", "√º", "y"]


def eingabe_und_test():
    """Fragt Eingaben ab und f√ºhrt alle drei Tests aus."""
    
    # 1Ô∏è‚É£ ist_gerade
    n = int(input("Geben Sie eine ganze Zahl ein: "))
    print(f"Ist {n} gerade? ‚Üí {ist_gerade(n)}")

    # 2Ô∏è‚É£ max2
    a = int(input("Erste ganze Zahl f√ºr den Vergleich: "))
    b = int(input("Zweite ganze Zahl f√ºr den Vergleich: "))
    print(f"Die gr√∂√üere Zahl ist: {max2(a, b)}")

    # 3Ô∏è‚É£ ist_vokal
    c = input("Geben Sie einen Buchstaben ein: ")
    print(f"Ist '{c}' ein Vokal? ‚Üí {ist_vokal(c)}")


# --- Funktionsaufruf ---
eingabe_und_test()

```

### Hinweise/L√∂sungsideen 5.2
- Bitten Sie ChatAI um drei kleine, kommentierte Funktionen, zum Beispiel:
  - `ist_gerade(n)` pr√ºft mit dem Modulo-Operator, ob `n` durch 2 teilbar ist.
  - `max2(a, b)` gibt die gr√∂√üere der beiden Zahlen zur√ºck.
  - `ist_vokal(c)` pr√ºft, ob ein einzelnes Zeichen ein Vokal ist.
- Fordern Sie eine kurze Erkl√§rung und einige Tests an.
- Beispiel: Die folgenden Funktionen zeigen eine m√∂gliche Umsetzung.

```python
def ist_gerade(n):
    # Gibt True zur√ºck, wenn n durch 2 teilbar ist
    return n % 2 == 0

def max2(a, b):
    # Gibt die gr√∂√üere von zwei Zahlen zur√ºck
    return a if a >= b else b

def ist_vokal(c):
    # Pr√ºft, ob ein Zeichen ein deutscher Vokal (a, e, i, o, u, √§, √∂, √º) ist
    return c.lower() in ["a", "e", "i", "o", "u", "√§", "√∂", "√º"]

# Tests
print(ist_gerade(4), ist_gerade(5))        # True False
print(max2(3, 7), max2(10, 10))            # 7 10
print(ist_vokal("A"), ist_vokal("x"))      # True False
```
