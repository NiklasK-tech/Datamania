```python
# --- Hilfsfunktionen ---

def get_display_text(category, value):
    display_texts = {
        "time": {"a": "1-2 Stunden pro Woche", "b": "2-3 Stunden pro Woche", "c": "3-5 Stunden pro Woche", "d": "5-8 Stunden pro Woche"},
        "interest": {"a": "Digitales", "b": "Coaching", "c": "Projektarbeit", "d": "Persönliche Weiterbildung"},
        "action": {"a": "Social Media Management", "b": "Menschen hier in Lüneburg helfen", "c": "Entwicklungszusammenarbeit", "d": "Initiative ergreifen"}
    }
    return display_texts[category][value]

# ------> Nested (Helper-)Function: Diese Hilfsfunktionen wurden kombiniert, um diese effizienter zu gestalten ohne Redundanzen:
# entscheidende Zeile:  "if any(user_answer in details[key] for user_answer in user_answers[key])" wird nun auf eine
# einmalige Überprüfung reduziert!

# any()	Mind. eine Bedingung erfüllt (Vergleich zu "ODER": logischer Operator)	„Wenn eine deiner Interessen passt!“
# all()	Alle Bedingungen erfüllt (Vergleich zu "UND": logischer Operator)

def calculate_results_and_criteria(initiatives, user_answers):
    results = {}
    criteria = {}
    for initiative, details in initiatives.items():
        match_count = 0
        matched_texts = []
        for key in user_answers:
            if any(user_answer in details[key] for user_answer in user_answers[key]): 
                match_count += 1
                matched_texts.extend(
                    [get_display_text(key, user_answer) for user_answer in user_answers[key] if user_answer in details[key]]
                )
        percentage = round((match_count / 3) * 100, 1)  # 3 Kategorien
        results[initiative] = percentage
        criteria[initiative] = matched_texts
    return results, criteria

# --- Hauptfunktionen ---

def initiativ_o_mat():
    print("Willkommen beim Initiativ-O-Mat!\n")
    print("Lass uns eine passende studentische Initiative für dich finden.\n")

    # --- 1. Fragen definieren ---
    time_question = (
        "Wie viel Zeit kannst du pro Woche für die Initiative aufbringen?\n"
        "a) 1-2 Stunden\n"
        "b) 2-3 Stunden\n"
        "c) 3-5 Stunden\n"
        "d) 5-8 Stunden\n"
    )

    interest_question = (
        "Welche Art der Arbeit interessiert dich? (Mehrere Optionen mit Komma trennen)\n"
        "a) Digitales\n"
        "b) Coaching\n"
        "c) Projektarbeit\n"
        "d) Persönliche Weiterbildung\n"
    )

    action_question = (
        "Was möchtest du tun? (Mehrere Optionen mit Komma trennen)\n"
        "a) Social Media Management\n"
        "b) Menschen hier in Lüneburg helfen\n"
        "c) Entwicklungszusammenarbeit\n"
        "d) Initiative ergreifen\n"
    )

    # --- 2. Antworten sammeln ---
    user_answers = {
        "time": [input(time_question).lower().strip()],
        "interest": [ans.strip() for ans in input(interest_question).lower().split(",")],
        "action": [ans.strip() for ans in input(action_question).lower().split(",")]
    }

    # --- 3. Initiativen und ihre Eigenschaften ---
    initiatives = {
        "Viva con Agua": {"time": ["a"], "interest": ["a", "c"], "action": ["a", "c"]},
        "Rock your life": {"time": ["b"], "interest": ["b", "d"], "action": ["b", "d"]},
        "Enactus": {"time": ["c"], "interest": ["a", "c"], "action": ["c", "d"]},
        "AISEC": {"time": ["d"], "interest": ["c", "d"], "action": ["d"]}
    }

    # --- 4. Ergebnisse und Kriterien berechnen ---
    results, matching_criteria_results = calculate_results_and_criteria(initiatives, user_answers)

    # --- 5. Ergebnisse ausgeben ---
    print("\nHier sind deine Ergebnisse:\n")
    for initiative, percentage in results.items():
        if percentage > 0:
            print(f"--- {initiative} ---")
            print(f"Übereinstimmungen: {', '.join(matching_criteria_results[initiative])}")
            print(f"Passgenauigkeit: {percentage}%\n")

# --- Programmstart ---
initiativ_o_mat()

# user_answers = ['a', 'b']
# details = {'key': ['a', 'c']}
```
